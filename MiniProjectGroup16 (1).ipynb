```python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('CleanedCrashData.csv')

# Define mapping for 'Weather' and 'Injury Severity'
weather_mapping = {'CLEAR': 0, 'CLOUDY': 1, 'RAIN': 2, 'SNOW': 3, 'FOG': 4, 'SEVERE WINDS': 5, 'SLEET': 6, 'BLOWING SNOW': 7, 'SEVERE CROSS WIND GATE': 8, 'BLOWING SAND, SOIL, DIRT': 9, 'SMOG': 10, 'FREEZING RAIN/DRIZZLE': 11, 'SEVERE STORMS': 13}
severity_mapping = {'NO APPARENT INJURY': 0, 'POSSIBLE INJURY': 1, 'NON-INCAPACITATING INJURY': 2, 'INCAPACITATING INJURY': 3, 'FATAL INJURY': 4}

# Apply mapping
df['Weather_encoded'] = df['Weather'].map(weather_mapping)
df['Injury_Severity_encoded'] = df['Injury Severity'].map(severity_mapping)

# Group by 'Weather_encoded' and calculate mean 'Injury_Severity_encoded'
grouped = df.groupby('Weather_encoded')['Injury_Severity_encoded'].mean().reset_index()

# Bar plot
plt.figure(figsize=(10, 5))
plt.bar(grouped['Weather_encoded'], grouped['Injury_Severity_encoded'], color='skyblue')
plt.title('Correlation between Weather and Injury Severity')
plt.xlabel('Weather')
plt.ylabel('Injury Severity')
plt.grid(True)
plt.show()
```

In the above code, we had gpt AI create a visualize to answer the question: "What correlation is there between the severity of injury and the weather conditions?" Chatgpt was guided to encode each weather condition and each injury severity level, so it could visualize this in quantitative form.
