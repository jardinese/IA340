```python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the dataframe is called df
df = pd.read_sql(sql, conn)

# Sort the dataframe by 'name' and 'year'
df.sort_values(by=['name', 'year'], inplace=True)

# Calculate year-over-year income growth rate for each state
df['income_growth'] = df.groupby('name')['income'].pct_change()

# Plotting the result
plt.figure(figsize=(12, 6))
for name in df['name'].unique():
    plt.plot(df[df['name'] == name]['year'], df[df['name'] == name]['income_growth'], label=name)

plt.xlabel('Year')
plt.ylabel('Annual Income Growth Rate')
plt.title('Annual Income Growth Rate by State')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
